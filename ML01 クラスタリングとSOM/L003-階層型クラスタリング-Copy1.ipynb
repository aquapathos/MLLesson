```python
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import pdist
from scipy.cluster import hierarchy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

X = np.array([x for x in set(np.round(100*np.random.rand(100)).astype('int'))])
X = X[:8]
X
```

```python
Y = [[x] for x in X]
Y
```

```python
Z = pdist(Y)
Z   # ZはYの要素のペアそれぞれの距離の配列
```

```python
pd = linkage(Z,method='ward') # ウォード法
dendro = dendrogram(pd,labels=X,truncate_mode='lastp')
```

```python
pd = linkage(Z,method='single') # 最短距離法
dendro = dendrogram(pd,labels=X,truncate_mode='lastp')
```

```python
pd = linkage(Z,method='average') # 群平均法
dendro = dendrogram(pd,labels=X,truncate_mode='lastp')
```

```python
pd = linkage(Z,method='centroid') # 重心法
dendro = dendrogram(pd,labels=X,truncate_mode='lastp')
```

# 課題　
1. 重心法を使ってD={0,2,5,9,12,14} のデンドログラムを作成せよ

```python
X=np.array([0,2,5,9,14,12])
Z = [[x] for x in X]
pd = linkage(pdist(Z),method='centroid') # 重心法
dendro = dendrogram(pd,labels=X,truncate_mode='lastp')
```

2. 二次元平面上に {(0,0), (0,1), (0,2),(4,0), (4,1), (4,2)}の６点の点集合がある． これらの点をward 法を使って階層クラスタリングせよ。

```python
L = ['A\n(0,2)', 'B\n(1,3)', 'C\n(0,2)','D\n(3,0)', 'E\n(3,3)', 'F\n(4,1)','G\n(6,1)','H\n(6,2)']
Z = [(0,2), (1,3), (2,0),(3,0), (3,3), (4,1),(6,1),(6,2)]
pd = linkage(pdist(Z),method='ward') # ウォード法
dendro = dendrogram(pd,labels=L,truncate_mode='lastp')
```

```python
L = ['A\n(0,0)', 'B\n(0,1)', 'C\n(1,2)','D\n(2,2)', 'E\n(2,4)', 'F\n(4,3)','G\n(3,4)','H\n(5,3)']
Z = [(0,0), (0,1), (1,2),(2,2), (2,4), (4,3),(3,4),(5,3)]
pd = linkage(pdist(Z),method='single') # 最短距離法
dendro = dendrogram(pd,labels=L,truncate_mode='lastp')
```
